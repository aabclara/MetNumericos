### MatPlotLib

- .title:Define o título principal do gráfico.
- .xlabel: Rótulo para o eixo horizontal.
- .ylabel: Rótulo para o eixo vertical.
- .legend(): Exibe a legenda

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Criando Dados com NumPy ---
# Gerar 100 pontos igualmente espaçados entre -2 e 2
x = np.linspace(-2, 2, 100)

# Definir uma função (exemplo: uma parábola f(x) = x^2)
y = x**2

# Definir outra função (exemplo: uma linha reta g(x) = 2x + 1)
y2 = 2 * x + 1

# --- 2. Plotagem ---
plt.figure(figsize=(8, 6)) # Define o tamanho da figura (largura, altura em polegadas)

# Plotar a primeira função (parábola)
plt.plot(x, y, color='blue', linestyle='-', linewidth=2, label= "f\(x\) \= x^2") # r'' para LaTeX

# Plotar a segunda função (reta)
plt.plot(x, y2, color='red', linestyle='--', linewidth=1.5, label="g\(x\) \= 2x \+ 1")

# --- 3. Personalização do Gráfico ---
plt.title('Plotagem Básica de Funções', fontsize=16) # Título do gráfico
plt.xlabel('Eixo X', fontsize=12) # Rótulo do eixo X
plt.ylabel('Eixo Y', fontsize=12) # Rótulo do eixo Y
plt.grid(True, linestyle=':', alpha=0.7) # Adiciona uma grade com estilo pontilhado e transparência
plt.axhline(0, color='black', linewidth=0.5) # Adiciona uma linha horizontal em y=0 (eixo X)
plt.axvline(0, color='black', linewidth=0.5) # Adiciona uma linha vertical em x=0 (eixo Y)
plt.legend(fontsize=10) # Exibe a legenda com as labels definidas em plt.plot
plt.tick_params(labelsize=10) # Ajusta o tamanho da fonte dos ticks

# Definir limites para os eixos para uma melhor visualização (opcional)
plt.xlim(-2.2, 2.2)
plt.ylim(-1, 5)

# --- 4. Exibir o Gráfico ---
plt.show()
```
