## Mais um exemplo de como encontrar o intervalo que contém as raízes da função graficamente

```python
# Importação das bibliotecas 
import numpy as np
import matplotlib.pyplot as plt

# Definição da Função
def f(x):
    return 5*x**3 - 11*x**2 - 14*x + 10

# Desenhar o gráfico - Teste 01
# Tamanho do gráfico
plt.figure(figsize=(10,6))
# Título do gráfico
plt.title("Gráfico de f(x) = 5x³ - 11x² - 14x + 10$")
# Título dos labels
plt.xlabel("x")
plt.ylabel("f(x)")
# Definição dos x e y
x = np.linspace(-100,100,5000)
y = f(x)
plt.plot(x,y)
# Adicionando o grid
plt.grid(True)
# Colocando o eixo x e y em negrito
plt.axvline(x=0, color='black', linestyle='-')
plt.axhline(y=0, color='black', linestyle='-')
# Mostrando o gráfico
plt.show()

# Desenhar o gráfico - Teste 02
# Tamanho do gráfico
plt.figure(figsize=(10,6))
# Título do gráfico
plt.title("Gráfico de f(x) = 5x³ - 11x² - 14x + 10$")
# Título dos labels
plt.xlabel("x")
plt.ylabel("f(x)")
# Definição dos x e y
x = np.linspace(-25,25,5000)
y = f(x)
plt.plot(x,y)
# Adicionando o grid
plt.grid(True)
# Colocando o eixo x e y em negrito
plt.axvline(x=0, color='black', linestyle='-')
plt.axhline(y=0, color='black', linestyle='-')
# Mostrando o gráfico
plt.show()

# Desenhar o gráfico - Teste 03
# Tamanho do gráfico
plt.figure(figsize=(10,6))
# Título do gráfico
plt.title("Gráfico de f(x) = 5x³ - 11x² - 14x + 10$")
# Título dos labels
plt.xlabel("x")
plt.ylabel("f(x)")
# Definição dos x e y
x = np.linspace(-2,4,5000)
y = f(x)
plt.plot(x,y)
# Adicionando o grid
plt.grid(True)
# Colocando o eixo x e y em negrito
plt.axvline(x=0, color='black', linestyle='-')
plt.axhline(y=0, color='black', linestyle='-')
# Mostrando o gráfico
plt.show()
 
# Desenhar o gráfico - Teste 04
# Tamanho do gráfico
plt.figure(figsize=(10,6))
# Título do gráfico
plt.title("Gráfico de f(x) = 5x³ - 11x² - 14x + 10$")
# Título dos labels
plt.xlabel("x")
plt.ylabel("f(x)")
# Definição dos x e y
x = np.linspace(0,1,5000)
y = f(x)
plt.plot(x,y)
# Adicionando o grid
plt.grid(True)
# Colocando o eixo x e y em negrito
plt.axvline(x=0, color='black', linestyle='-')
plt.axhline(y=0, color='black', linestyle='-')
# Mostrando o gráfico
plt.show()

# Desenhar o gráfico - Teste 05
# Tamanho do gráfico
plt.figure(figsize=(10,6))
# Título do gráfico
plt.title("Gráfico de f(x) = 5x³ - 11x² - 14x + 10$")
# Título dos labels
plt.xlabel("x")
plt.ylabel("f(x)")
# Definição dos x e y
x = np.linspace(0.4,0.6,5000)
y = f(x)
plt.plot(x,y)
# Adicionando o grid
plt.grid(True)
# Colocando o eixo x e y em negrito
plt.axvline(x=0, color='black', linestyle='-')
plt.axhline(y=0, color='black', linestyle='-')
# Mostrando o gráfico
plt.show()
```
