A equação $f(x) = x^5 - 6x^4 - 14x^3 +72x^2+44x-180$ possui três raízes positivas no intervalo [0,7].
1. Determine essas raízes pelos métodos $\\$
    a. Bisseção $\\$
    b. Secante $\\$
    c. Falsa Posição

```python
# Importação do arquivo com os scripts padrão
import acsa_silveira as acsa

# Definição da função f(x)
def f(x):
        return x**5 - 6*x**4 - 14*x**3 + 72*x**2 + 44*x - 180
```

```python
# LETRA A
acsa.bissecao(0,7,0.00009,f)
```

```python
# LETRA B
# Método da secante com 100 iterações e tolerância de 1e-6
acsa.secante(f,0,7)
```

```python
# LETRA C
# Método da falsa posição com 100 iterações e tolerância de 1e-6
acsa.falsa_posicao(f,0,7)
```

2. Compare graficamente a convergência dos métodos usados.

```python
# Comparação das funções
# Bisseção - rosa
# Secante - red 
# Falsa Posição - blue
acsa.grafico_comparativo(f,0,7,0.00009)
```
