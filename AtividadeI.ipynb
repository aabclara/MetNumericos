**A função $f(x) = x^5 -6x^4 - 14x^3 + 73x^2 + 44x - 180$ possui três raízes positivas no intervalo [0, 7].**

```python
# importando meu help
import MariaClara as help
import numpy as np
import matplotlib.pyplot as plt
```

```python
# criando uma função f(x) para facilitar o desenvolvimento:
def f(x):
    return x**5 - 6*(x**4) - 14*(x**3) + 73*(x**2) + 44*x - 180
```

#### *Questão I:*
Determine essas raízes pelos métodos da bissecção, secante e falsa posição:

```python
help.grafico(f, 0, 7, 1000, "Gráfico de $x**5 - x**5 - 6*(x**4) - 14*(x**3) + 73*(x**2) + 44*x - 180$")
```

Após a plotagem e interpretação do gráfico acima, podemos concluir que as raízes de f estão dentro dos intervalos:
- [1, 2]
- [2, 3]
- [6, 7]

```python
help.bissecao(0, 7,0.00009,f)
help.bissecao(1, 2,0.00009,f)
help.bissecao(3, 2,0.00009,f)
help.bissecao(6, 7,0.00009,f)
```

```python
raiz1_secante = help.secante(f, 1, 2)
print(raiz1_secante)

raiz2_secante = help.secante(f, 3, 2)
print(raiz2_secante)

raiz3_secante = help.secante(f, 6, 7)
print(raiz3_secante)
```

```python
raiz1_secante = help.falsa_posicao(f, 1, 2)
print(raiz1_secante)

raiz2_secante = help.falsa_posicao(f, 3, 2)
print(raiz2_secante)

raiz3_secante = help.falsa_posicao(f, 6, 7)
print(raiz3_secante)
```
