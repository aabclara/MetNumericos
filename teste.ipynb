```python
def Secante(f,x0, x1, tol = 1e-6, max_iter = 100):
    for i in range(max_iter):
        x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        if abs(x2 - x1) < tol:
            return x2
        x0 = x1
        x1 = x2
    return None # se nÃ£o converge
```

```python
def falsaPosicao(f, a, b, tol = 1e-6, max_iter = 100):
    """
    Encontra a raiz, no intervalo [x0, x1], da equaÃ§Ã£o definida em f.
    Parada: no mÃ¡ximo, max_iter, iteraÃ§Ãµes ou diferenÃ§a entre os limites do intervalo menor que tol.
    """
    if f(a) * f(b) >= 0:
        print("Erro! f(a) e f(b) devem ter sinais opostos!")
        return None
    
    for i in range(max_iter):
        c = (a * f(b) - b*f(a)) / (f(b)-f(a))
        if abs(f(c)) < tol:
            return c
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
        return None #nÃ£o converge
    return c
```
