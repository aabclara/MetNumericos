1. Encontre uma aproximação para a raiz da função abaixo dentro do intervalo 
[1, 2][1,2], com precisão de $10^{−4}$:  
    $f(x)=x^3−2x−5$

```python
def f(x):
    return x**3 - 2*x - 5
MariaClaraM.secante(f, 2, 1, 10**(-4))
```

2. A função $f(x)=cos(x)−x$ possui uma raiz entre 0 e 1. Utilize um método adequado para aproximá-la com tolerância de 
$10^{-5}$.

```python
def f(x):
    return np.cos(x) - x

MariaClaraM.bissecao(f, 0, 1, 10**(-5))
```
